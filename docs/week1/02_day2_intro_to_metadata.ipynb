## One tiny dataset “story”

We will attach three simplified metadata types to this dataset:
- **Descriptive metadata**: what the dataset is about (title, description).
- **Structural metadata**: how the dataset is organized (column meaning).
- **Administrative metadata**: access and management details (license).


```python
descriptive_metadata = {
    "title": "2025 Global Climate Data",
    "description": "Fictional annual climate summaries for two countries from 2021 to 2023.",
    "keywords": ["climate", "temperature", "rainfall", "CO2 emissions"],
}

structural_metadata = {
    "rows": "country-year observations",
    "columns": {
        "country": "fictional country name",
        "year": "year of observation",
        "avg_temperature": "average temperature in degrees Celsius",
        "rainfall_mm": "annual rainfall in millimeters",
        "co2_emissions": "CO2 emissions in fictional units",
    },
}

administrative_metadata = {
    "license": "CC BY 4.0 (example)",
    "version": "1.0 (teaching dataset)",
    "contact": "research-data@example.org (example)",
}

descriptive_metadata, structural_metadata, administrative_metadata
```

## How metadata supports FAIR (especially Findable + Reusable)

- **Findable**: descriptive metadata (title/description/keywords) helps discovery.
- **Reusable**: structural + administrative metadata (column meaning + license) helps others use the data correctly.
