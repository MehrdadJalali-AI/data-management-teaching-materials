This notebook introduces the concept of **metadata** using a small teaching dataset.  
It is designed for **classroom presentation**, so each step is short, visual, and easy to explain.

- understand what metadata is

```python
# Notebook bootstrap: download required files from GitHub if missing
from pathlib import Path
from urllib.request import urlretrieve

GITHUB_RAW_BASE = "https://raw.githubusercontent.com/MehrdadJalali-AI/data-management-teaching-materials/main"

required_files = [
    ("data/raw/climate_sample.csv", "data/raw/climate_sample.csv"),
]

for local_rel, github_rel in required_files:
    local_path = Path(local_rel)
    if not local_path.exists():
        local_path.parent.mkdir(parents=True, exist_ok=True)
        url = f"{GITHUB_RAW_BASE}/{github_rel}"
        urlretrieve(url, local_path)
        print(f"Downloaded: {local_path}")

print("Bootstrap complete.")
```

```python
# Step 0 — Load the dataset
# Keep the original repository file path.
import pandas as pd
from pathlib import Path
from IPython.display import display, Markdown

repo_root = Path().resolve()
csv_path = repo_root / "data" / "raw" / "climate_sample.csv"

print("Expected CSV path:", csv_path)
print("File exists:", csv_path.exists())

df = pd.read_csv(csv_path)
display(df)
```

It contains values such as country, year, temperature, rainfall, and CO₂ emissions.

```python
# Step 2 — Quick inspection of the dataset

# Shows the number of rows and columns (rows, columns)
# Helps quickly understand dataset size
print("Shape of dataset:", df.shape)

print("\nColumns:")
# Lists all column names in the dataset
# Useful to quickly scan available features
for col in df.columns:
    print("-", col)

print("\nData types:")
# Displays the data type of each column (e.g., int, float, object, datetime)
# Important for identifying:
# - numeric vs categorical features
# - potential type issues (e.g., numbers stored as strings)
display(df.dtypes.to_frame("dtype"))

print("\nBasic summary:")

# Generates a comprehensive summary of ALL columns in the dataset
# include="all" ensures both numeric and non-numeric (categorical/text) columns are included
# For numeric columns → shows stats like mean, std, min, max
# For categorical columns → shows unique values, most frequent value (top), and its frequency
#
# transpose() flips the table so that:
# - each row represents a column (feature)
# - each column represents a statistic
# → This makes the output much more readable and easier to compare features
#
# display() is used (instead of print) for a clean, nicely formatted table in Jupyter Notebook
display(df.describe(include="all").transpose())
```

## Step 2 — Why raw data alone is not enough

Even if the dataset is small and readable, several questions remain:

- Is this a real or fictional dataset?
- What exactly does `avg_temperature` mean?
- What unit is used for `co2_emissions`?
- Who created the dataset?
- Can others reuse it legally?
- What does one row represent?

These answers are usually not stored in the data values themselves.  
They are provided by **metadata**.


## Step 3 — Define three metadata types

For teaching purposes, we will describe the dataset using three simple metadata layers:

### 1. Descriptive metadata
Helps people **find** and understand what the dataset is about.

### 2. Structural metadata
Explains how the dataset is **organized**.

### 3. Administrative metadata
Describes management, ownership, access, versioning, and reuse conditions.


```python
# Step 3A — Descriptive metadata

# A dictionary storing high-level information ABOUT the dataset (not the data itself)
# This is essential for FAIR principles (Findability & Reusability)
descriptive_metadata = {
    "title": "2025 Global Climate Data",  # Name of the dataset
    "description": "Fictional annual climate summaries for two countries from 2021 to 2023.",  # What the dataset contains
    "keywords": ["climate", "temperature", "rainfall", "CO2 emissions"],  # Searchable terms for discovery
    "creator": "Teaching repository example",  # Who created the dataset
    "subject": "Climate and environmental indicators"  # General topic/domain
}

# Convert metadata dictionary into a structured table (DataFrame)
# This makes metadata easier to read, display, and export if needed
display(pd.DataFrame(
    {
        "Field": list(descriptive_metadata.keys()),  # Metadata field names (e.g., title, description)
        "Value": [str(v) for v in descriptive_metadata.values()]  # Corresponding values (converted to string for consistency)
    }
))

# Result:
# A clean two-column table:
# Field   | Value
# --------|----------------------
# title   | 2025 Global Climate Data
# ...
#
# This format is useful for:
# - documentation
# - sharing metadata in reports
# - preparing for standards like DataCite or schema.org
```

```python
# Step 3B — Structural metadata
structural_metadata = {
    "rows": "Each row represents one country-year observation.",
    "columns": {
        "country": "Country name",
        "year": "Year of observation",
        "avg_temperature": "Average temperature in degrees Celsius",
        "rainfall_mm": "Annual rainfall in millimeters",
        "co2_emissions": "CO2 emissions in fictional units",
    },
}

structural_rows = [{"Column": k, "Meaning": v} for k, v in structural_metadata["columns"].items()]
display(pd.DataFrame(structural_rows))
```

```python
# Step 3C — Administrative metadata
administrative_metadata = {
    "license": "CC BY 4.0 (example)",
    "version": "1.0 (teaching dataset)",
    "contact": "research-data@example.org (example)",
    "access": "Open for teaching purposes",
    "provenance_note": "Created as a fictional classroom dataset"
}

display(pd.DataFrame(
    {"Field": list(administrative_metadata.keys()),
     "Value": [str(v) for v in administrative_metadata.values()]}
))
```

## Step 4 — Data vs. Metadata

A simple way to explain it in class:

- **Data** = the actual values  
  Example: `Germany, 2023, 14.2, 620, 11.3`

- **Metadata** = information about the data  
  Example:  
  - title of the dataset  
  - meaning of each column  
  - license  
  - creator  
  - version

### Teaching analogy
A dataset is like a **book**.  
Metadata is like the **catalog entry** and **front matter** that tell us:
- title
- author
- subject
- edition
- access conditions


```python
# Step 5 — Combine metadata into one classroom view
metadata_overview = [
    {"Type": "Descriptive", "Purpose": "Discovery and understanding", "Example": "Title, description, keywords"},
    {"Type": "Structural", "Purpose": "Organization and interpretation", "Example": "Rows, columns, units"},
    {"Type": "Administrative", "Purpose": "Management and reuse", "Example": "License, version, contact"},
]

display(pd.DataFrame(metadata_overview))
```

## Step 6 — Why metadata matters for FAIR

Metadata is essential for FAIR data practices.

### Findable
A dataset becomes easier to find when it has:
- a clear title
- useful keywords
- a good description
- ideally a persistent identifier

### Reusable
A dataset becomes easier to reuse when it has:
- clear column meanings
- units and structure
- license information
- version and provenance details


```python
# Step 6 — Connect our example metadata to FAIR
fair_mapping = [
    {"FAIR principle": "Findable", "Helpful metadata fields": "title, description, keywords, creator"},
    {"FAIR principle": "Reusable", "Helpful metadata fields": "column meanings, license, version, provenance"},
]

display(pd.DataFrame(fair_mapping))
```
